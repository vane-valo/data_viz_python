## Bar Chart

#### First we read our dataset with pd.read_csv() and we save it on a variable name df_used_cars
#### We use the function 'bar' to create the bar chart. The x parameter stands for the x coordinate (horizontal values). The height parameters indicates the height of the bars. The color parameter, we can choose a color name from a predefined color list available in the seaborn library.

```python
path="/Users/vanessavaca/Python/Data_Analyst_Python/Data_Viz/Data_Viz_Notebook/data/bar_chart_data.csv"
df_used_cars = pd.read_csv(path)
```

## Pie Chart

#### First we read our dataset with pd.read_csv() and we save it on a variable name df_fuel_engine_type
#### We use the function 'pie' to create the pie chart. We set the row of the dataset as parameter. The 'labels' parameter to include pie chart labels. The 'autopct' takes a string or function to label the wedges. The 'textprops' is a dictionary of arguments for the text object.
#### We'll alter seaborn deafult color palette and set it to 'colorblind'

```python
path="/Users/vanessavaca/Python/Data_Analyst_Python/Data_Viz/Data_Viz_Notebook/data/pie_chart_data.csv"
df_fuel_engine_type = pd.read_csv(path)
```

```python
df_fuel_engine_type
```

```python
plt.figure(figsize=(10,8))
plt.pie(df_fuel_engine_type['Number of Cars'],
       labels=df_fuel_engine_type['Engine Fuel Type'].values,
       autopct='%.2f%%',
       textprops={'size':'x-large',
                 'fontweight':'bold',
                 'rotation':'30',
                 'color':'w'})

plt.legend()
plt.title('Cars bye Engine Fuel Type', fontsize=18, fontweight='bold')
plt.show()
```

## Stacked Area Chart

#### First we read our dataset with pd.read_csv() and we save it on a variable name df_fuel_engine_type
#### We use the function 'stackplot' to create a stacked area chart. We set the first parameter as the x-axis. The next parameters are the 'labels'.
#### set_style() from seaborn allows us to pick a background style: white, whitegrid, dark, etc.
#### despine() from seaborn allows us to remove the top and right border of the chart.

```python
path="/Users/vanessavaca/Python/Data_Analyst_Python/Data_Viz/Data_Viz_Notebook/data/stacked_area_chart_data.csv"
df_fuel_engine_type = pd.read_csv(path)
```

## Scatter Plot

#### First we read our dataset with pd.read_csv() and we save it on a variable named df_real_state
#### We use the function 'scatter' to create a scatter plot. We set the first parameter as the x-axis, and the second parameter as the y-axis.
#### The 'alpha' parameter controls the transparency level and takes values from 0 to 1.
#### The 'c' parameter for the color code according to the variable.
#### Having a large data set might result in overplotting. To avoid this we can add a transparency level.
#### In scatter plot we can color code according to a third variable.

```python
path="/Users/vanessavaca/Python/Data_Analyst_Python/Data_Viz/Data_Viz_Notebook/data/scatter_data.csv"
df_real_state = pd.read_csv(path)
df_real_state.head()
```

```python
plt.figure(figsize=(12,8))
scatter = plt.scatter(df_real_state['Area (ft.)'],
           df_real_state['Price'],
           alpha=0.6,
           c=df_real_state['Building Type'],
           cmap='viridis')
plt.legend(*scatter.legend_elements(),
          loc="upper left",
          title="Buildying Type")
plt.title("Relationship between Area and Price of California Real Estate",
         fontsize=14,
         weight='bold')
plt.xlabel("Area (sq. ft)", weight="bold")
plt.ylabel("Price (000's of $)")

plt.show()
```

#### We can make a scatter plot with the seaborn library using the function 'scatterplot'
#### The first parameter is the x-axis and the second one the y-axis.
#### sns.scatterplot uses the 'hue' parameter to color code on a third variable.
#### 'palette' is used to pick the colors for the third variable.
#### 's' parameter let us specify the size of each point on the scatter.

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=df_real_state['Price'],
               y=df_real_state['Area (ft.)'],
               hue=df_real_state['Building Type'],
               palette=['black', 'darkblue', 'purple', 'pink', 'white'],
               s=100)
plt.title("Relationship between Area and Price of California Real Estate",
         fontsize=14,
         weight='bold')
plt.xlabel("Area (sq. ft)", weight="bold")
plt.ylabel("Price (000's of $)")

plt.show()
```

## Regression Plot

#### First we read our dataset with pd.read_csv() and we save it on a variable named df_ad_expenditure
#### We'll rely on the 'seaborn' library to incorporate regression lines onto a scatter plot.
#### There are two functions we can use, sns.regplot() and sns.lmplot()
#### The difference between two is that sns.regplot() accepts a variety of formats for the x and y variables such as numpy arrays, pandas series, etc. And sns.lmplot() can only support x and y in the form of strings also known as 'long form' or 'tidy' data. , that means that it supports a wider range of features.
#### We begin with the sns.regplot() The parameter are specified for the x-axis, y-axis and the dataframe where you are taking the variables.
#### 'scatter_kws' is to change the color of the points and 'line_kws' is to change the color of the line.
#### For sns.lmplot() we specified as well the x, y and data variables. The size is controlled through the 'height' parameter.

```python
path="/Users/vanessavaca/Python/Data_Analyst_Python/Data_Viz/Data_Viz_Notebook/data/scatter_plot_ii.csv"
df_ad_expenditure = pd.read_csv(path)
df_ad_expenditure.head()
```

```python
#plt.figure(figsize=(10,8))
sns.set(rc={'figure.figsize':(9,6)})
sns.regplot(x="Budget",
           y="Sales",
           data=df_ad_expenditure,
           scatter_kws={'color':'k'},
           line_kws={'color':'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, weight="bold")

plt.show()
```

```python
sns.lmplot(x='Budget',
          y='Sales',
          data=df_ad_expenditure,
          height=10,
          scatter_kws={'color':'k'},
          line_kws={'color':'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, weight="bold")

plt.show()
```

## Bar and Line Chart

#### First we read our dataset with pd.read_csv() and we save it on a variable named df_kdnuggets
#### The PercentFormatter will allow us to change the scale of the secondary y-axis to percentages.
#### The figure 'axes' are different from the chart's x- and y-axis. A figure encompases all chart elements. A figure can have multiple axes, we'll be using two. The first axes will be the bar chart, and the second axes will be the line chart.
#### We'll use ax.bar() where for parameteres will specify the x- and y- axis for the bar chart.
#### For the second axes, we'll create a variable containing the first one, 'ax1', where the function ax.twinx() is to create a twin Axes sharing the x-axis.
#### We need to clearly label the axis to signify we're displaying percentages.
#### set_ylim() let us alter the y-axis range by adding a min and max value.

```python
path="/Users/vanessavaca/Python/Data_Analyst_Python/Data_Viz/Data_Viz_Notebook/data/bar_line_chart_data.csv"
df_kdnuggets = pd.read_csv(path)
df_kdnuggets
```

```python
fig, ax = plt.subplots(figsize=(10,7))

ax.bar(df_kdnuggets['Year'],
      df_kdnuggets['Participants'],
      color="k")
ax.set_ylabel("Number of Participants",
         weight="bold")
ax.tick_params(axis="y",
              width=2,
              labelsize="large")
ax1 = ax.twinx()
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1.0))
ax1.plot(df_kdnuggets['Year'],
        df_kdnuggets['Python Users'],
        color="#b60000",
        marker="D")
ax1.set_ylabel("Python Users",
              color="#b60000",
              weight="bold")
ax1.tick_params(axis="y",
               colors="#b60000",
               width=2,
               labelsize="large")
ax.set_title("KD Nuggets Survey Python Users (2012-2019)",
            fontsize=14,
            weight="bold")
```
